```python
print("\nAttempting to convert 'Order Date' to datetime...")

# We use 'format' to tell pandas exactly how to read your date strings.
# %d for day, %m for month, %Y for 4-digit year.
# 'errors='coerce' will turn any date it CANNOT read into NaT (Not a Time).
# This is a good safety net.
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y', errors='coerce')

print("Conversion attempt complete.")

# --- Verify the data type *after* this conversion ---
print("\nData type of 'Order Date' AFTER conversion:")
print(df['Order Date'].dtype)

# --- Check for any dates that couldn't be converted ---
# If any dates were unreadable, they'll now be NaT (Not a Time).
# Let's see how many there are.
num_unparsed = df['Order Date'].isnull().sum()
if num_unparsed > 0:
    print(f"\nWarning: {num_unparsed} dates in 'Order Date' could not be understood and became 'NaT'.")
    print("These rows might need further inspection or should be dropped.")
    # Show a few rows that failed conversion
    print(df[df['Order Date'].isnull()].head())
    # For now, let's remove these problematic rows so the rest of the analysis works
    df.dropna(subset=['Order Date'], inplace=True)
    print(f"{num_unparsed} rows with unreadable dates have been removed.")
else:
    print("\nAll 'Order Date' values were successfully converted!")
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')
df = df.dropna(subset=['Order Date', 'Sales'])

# ----- 2. Extract year -----
df['Year'] = df['Order Date'].dt.year

# ----- 3. Group by year -----
yearly_sales = df.groupby('Year')['Sales'].sum().reset_index()

# ----- 4. Plot -----
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(data=yearly_sales, x='Year', y='Sales')

plt.title("Total Sales per Year")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ----- 1. Ensure 'Order Date' is datetime (you already did this, but safe to re-run) -----
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df = df.dropna(subset=['Order Date'])

# ----- 2. Convert Sales to numeric -----
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')
df = df.dropna(subset=['Sales'])

# ----- 3. Extract Year as an integer -----
df['Year'] = df['Order Date'].dt.year.astype(int)

# ----- 4. Group by Year -----
yearly_sales = df.groupby('Year', as_index=False)['Sales'].sum()

# ----- 5. Plot -----
plt.figure(figsize=(10, 6))
sns.barplot(data=yearly_sales, x='Year', y='Sales')

plt.title("Total Sales Per Year")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.tight_layout()
plt.show()
```

```python
country_sales = df.groupby('State', as_index=False)['Sales'].sum()

plt.figure(figsize=(12, 8))
sns.barplot(data=country_sales, x='Sales', y='State', palette='viridis')

plt.title("Total Sales per State")
plt.xlabel("Total Sales")
plt.ylabel("States")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
            
```

```python
region_sales = df.groupby('Region', as_index=False)['Sales'].sum()

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(data=region_sales, x='Region', y='Sales')

plt.title("Total Sales per Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.tight_layout()
plt.show()
```

```python
category_sales = df.groupby('Category', as_index=False)['Sales'].sum()


plt.figure(figsize=(10, 6))
sns.barplot(data=category_sales, x='Category', y='Sales')

plt.title("Total Sales per Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.tight_layout()
plt.show()
```

```python
product_sales = df.groupby('Product Name', as_index=False)['Sales'].sum()

# Sort so the biggest products appear first
product_sales = product_sales.sort_values(by='Sales', ascending=False)

# Plot â€” shows top 20 to avoid clutter
plt.figure(figsize=(12, 8))
sns.barplot(data=product_sales.head(20), x='Sales', y='Product Name')

plt.title("Top 20 Products by Total Sales")
plt.xlabel("Total Sales")
plt.ylabel("Product Name")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Step 1: Sum sales per State + Product
state_product_sales = (
    df.groupby(['State', 'Product Name'], as_index=False)['Sales']
      .sum()
)

# Step 2: Get the top product for each state
highest_product_per_state = (
    state_product_sales.sort_values(['State', 'Sales'], ascending=[True, False])
                       .groupby('State')
                       .head(1)
                       .reset_index(drop=True)
)

highest_product_per_state
```

```python
plt.figure(figsize=(14, 18))

# Sort by sales to make the chart look cleaner
sorted_state_sales = highest_product_per_state.sort_values(by='Sales', ascending=True)

sns.barplot(
    data=sorted_state_sales,
    x='Sales',
    y='State',
    hue='Product Name',
    dodge=False
)

plt.title("Top Selling Product in Each State", fontsize=16)
plt.xlabel("Total Sales")
plt.ylabel("State")
plt.tight_layout()
plt.show()
```
