```python
#transform the total charges column to numeric
customer_df['TotalCharges'] = pd.to_numeric(customer_df['TotalCharges'], errors='coerce')
customer_df.info()
```

```python
#Check for missing values in TotalCharges
customer_df['TotalCharges'].isna().sum()
```

```python
#fix the missing totalCharges with 0
customer_df['TotalCharges'].fillna(0, inplace=True)
```

```python
customer_df['TotalCharges'].isna().sum()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=customer_df, x='Churn', y='TotalCharges')
plt.title('Total Charges vs Churn')
plt.xlabel('Churn')
plt.ylabel('Total Charges')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=customer_df, x='Churn', y='TotalCharges')
plt.title("Total Charges vs Churn (Violin Plot)")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=customer_df, x='TotalCharges', hue='Churn', kde=True, bins=30)
plt.title("Distribution of Total Charges by Churn")
plt.xlabel("Total Charges")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=customer_df, x='Churn', y='MonthlyCharges')
plt.title('Monthly Charges vs Churn')
plt.xlabel('Churn')
plt.ylabel('Monthly Charges')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=customer_df, x='gender', hue='Churn')
plt.title("Gender vs Churn Count")
plt.xlabel("Gender")
plt.ylabel("Number of Customers")
plt.show()
```

```python
gender_churn_rate = (
    customer_df.groupby('gender')['Churn']
    .value_counts(normalize=True)
    .mul(100)
    .rename('percentage')
    .reset_index()
)

plt.figure(figsize=(8, 5))
sns.barplot(data=gender_churn_rate, x='gender', y='percentage', hue='Churn')
plt.title("Churn Rate (%) by Gender")
plt.ylabel("Percentage (%)")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=customer_df, x='TechSupport', hue='Churn')
plt.title("TechSupport vs Churn Count")
plt.xlabel("TechSupport")
plt.ylabel("Number of Customers")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=customer_df, x='PhoneService', hue='Churn', palette='Set2')
plt.title('Phone Service vs Churn')
plt.xlabel('Phone Service')
plt.ylabel('Number of Customers')
plt.show()
```

```python
# Pivot table for churn percentage
churn_contract = pd.crosstab(customer_df['Contract'], customer_df['Churn'])
churn_contract['Total'] = churn_contract.sum(axis=1)
churn_contract['Churn_Percentage'] = (churn_contract['Yes'] / churn_contract['Total']) * 100
churn_contract
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x=churn_contract.index, y=churn_contract['Churn_Percentage'], palette='Set1')
plt.title('Churn Percentage by Contract Type')
plt.ylabel('Churn Percentage (%)')
plt.show()
```

```python
# Crosstab for churn percentage
churn_security = pd.crosstab(customer_df['OnlineSecurity'], customer_df['Churn'])
churn_security['Total'] = churn_security.sum(axis=1)
churn_security['Churn_Percentage'] = (churn_security['Yes'] / churn_security['Total']) * 100
churn_security

plt.figure(figsize=(6,4))
sns.barplot(x=churn_security.index, y=churn_security['Churn_Percentage'], palette='Set1')
plt.title('Churn Percentage by Online Security')
plt.ylabel('Churn Percentage (%)')
plt.show()

```

```python
churn_tvStreaming = pd.crosstab(customer_df['StreamingTV'], customer_df['Churn'])
churn_tvStreaming['Total'] = churn_tvStreaming.sum(axis=1)
churn_tvStreaming['Churn_Percentage'] = (churn_tvStreaming['Yes'] / churn_tvStreaming['Total']) * 100
churn_tvStreaming

plt.figure(figsize=(6,4))
sns.barplot(x=churn_tvStreaming.index, y=churn_tvStreaming['Churn_Percentage'], palette='Set1')
plt.title('Churn Percentage by Streaming TV')
plt.ylabel('Churn Percentage (%)')
plt.show()
```
