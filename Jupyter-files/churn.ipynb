```python
#transform the total charges column to numeric
customer_df['TotalCharges'] = pd.to_numeric(customer_df['TotalCharges'], errors='coerce')
customer_df.info()
```

```python
#Check for missing values in TotalCharges
customer_df['TotalCharges'].isna().sum()
```

```python
customer_df[customer_df['TotalCharges'].isna()][['customerID','tenure','TotalCharges']]
```

```python
#fix the missing totalCharges with 0
customer_df['TotalCharges'].fillna(0, inplace=True)
```

```python
customer_df['TotalCharges'].isna().sum()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=customer_df, x='Churn', y='TotalCharges')
plt.title('Total Charges vs Churn')
plt.xlabel('Churn')
plt.ylabel('Total Charges')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=customer_df, x='Churn', y='TotalCharges')
plt.title("Total Charges vs Churn (Violin Plot)")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=customer_df, x='TotalCharges', hue='Churn', kde=True, bins=30)
plt.title("Distribution of Total Charges by Churn")
plt.xlabel("Total Charges")
plt.ylabel("Count")
plt.show()
```
